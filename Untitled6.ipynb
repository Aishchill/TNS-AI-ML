```python
import pandas as pd
import matplotlib.pyplot as plt

# Create synthetic time series data
dates = pd.date_range('2024-01-01', periods=12, freq='M')

sales = [120, 140, 135, 150, 170, 190,
         180, 210, 230, 220, 240, 260]

# Line plot
plt.figure(figsize=(10,5))
plt.plot(dates,
         sales,
         marker='o',
         linestyle='-',
         color='green',
         label='Sales')

# Peak point
peak_index = sales.index(max(sales))

plt.annotate('Peak Sales',
             xy=(dates[peak_index], sales[peak_index]),
             xytext=(dates[peak_index], sales[peak_index]+20),
             arrowprops=dict(arrowstyle='->'))

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)

plt.show()
```
